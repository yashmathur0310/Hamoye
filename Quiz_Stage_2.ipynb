```python
import pandas as pd 
import numpy as np 
```

```python
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('energydata_complete.csv')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df=df.drop(['date','lights'],axis=1)
```

```python
columns=df.columns
```

```python
df=pd.DataFrame(scaler.fit_transform(df))
df.columns=columns
df
```

```python
x_12=df[['T2']].values
y_12=df['T6'].values
```

```python
from sklearn.model_selection import train_test_split
x_12_train,x_12_test,y_12_train,y_12_test=train_test_split(x_12,y_12,random_state=42,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
regressor_12=LinearRegression()
regressor_12.fit(x_12_train,y_12_train)
```

```python
y_pred_12=regressor_12.predict(x_12_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_12_test,y_pred_12)
np.round(r2,2)
```

##### 0.64 R^2 score

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_12_test,y_pred_12)
np.round(mse,2)
```

##### 0.01 mse 

```python
rss=np.sum(np.square(y_12_test-y_pred_12))
np.round(rss,2)
```

##### 66.12 rss 


```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_12_test,y_pred_12))
np.round(rmse,3)
```

##### 0.106 rmse
